```python
# Merge the plays and games dataframes
full_play_df = pd.merge(plays_df, games_df, on = 'gameId', how = 'inner')
full_play_df.head()
```

```python
# Add the defense column identifying the defensive team
full_play_df['def'] = np.where(full_play_df['possessionTeam'] != full_play_df['homeTeamAbbr'], 
                               full_play_df['homeTeamAbbr'], full_play_df['visitorTeamAbbr'])
```

```python
# The resulting merged full play dataframe
full_play_df.head()
```

```python
# The number of defensive plays for each team
count_df = full_play_df.groupby('def')['playId'].count()
count_df
```

```python
# The number of completions for each team
completion_df = full_play_df.groupby('def')['passResult'].apply(lambda x: (x=='C').sum())
completion_df
```

```python
# The number of incompletions for each team
incompletion_df = full_play_df.groupby('def')['passResult'].apply(lambda x: (x=='I').sum())
incompletion_df
```

```python
# The number of sacks for each team
sack_df = full_play_df.groupby('def')['passResult'].apply(lambda x: (x=='S').sum())
sack_df
```

```python
# The number of interceptions for each team
interception_df = full_play_df.groupby('def')['passResult'].apply(lambda x: (x=='IN').sum())
interception_df
```

```python
# The sum of the epa's for each team
epa_df = full_play_df.groupby('def')['epa'].sum()
epa_df
```

```python
# The merged dataframe for the initial stats
merge_list = [count_df, completion_df, incompletion_df, sack_df, interception_df, epa_df]
team_df = pd.concat(merge_list, axis=1)
team_df.columns = ['totalPassAttempts', 'completions', 'incompletions', 'sacks', 'interceptions', 'epa']
team_df.reset_index()
```

```python
# The team completion percentage
team_df['completionPct'] = team_df['completions'] / team_df['totalPassAttempts']
team_df.reset_index()
```

```python
# The team completion rank
team_df['completionRank'] = team_df['completionPct'].rank(ascending=1)
team_df.reset_index()
```

```python
# The team epa rank
team_df['epaRank'] = team_df['epa'].rank(ascending=1)
team_df.reset_index()
```

```python
# The team sack rank
team_df['sackRank'] = team_df['sacks'].rank(ascending=0)
team_df.reset_index()
```

```python
# The team interception rank
team_df['intRank'] = team_df['interceptions'].rank(ascending=0)
team_df.reset_index()
```

```python
# The team combined rank - equal weighting for each ranking
team_df['combinedRank'] = team_df['completionRank'] + team_df['epaRank'] + team_df['sackRank'] + team_df['intRank']
team_df = team_df.sort_values(by=['combinedRank'])
team_df.reset_index()
```

```python
# The team final rank
team_df['finalRank'] = team_df['combinedRank'].rank(ascending=1)
team_df = team_df.sort_values(by=['finalRank'])
team_df.reset_index()
```
