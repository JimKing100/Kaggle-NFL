```python
# Reduce the weeks data to pass_outcome data
def week_reduce(df):
    df = df[(df['event'] == 'pass_outcome_caught') | 
            (df['event'] == 'pass_outcome_incomplete') |
            (df['event'] == 'pass_outcome_interception') |
            (df['event'] == 'pass_outcome_touchdown')]
    return df

week1_df = week_reduce(week1r_df)
week2_df = week_reduce(week2r_df)
week3_df = week_reduce(week3r_df)
week4_df = week_reduce(week4r_df)
week5_df = week_reduce(week5r_df)
week6_df = week_reduce(week6r_df)
week7_df = week_reduce(week7r_df)
week8_df = week_reduce(week8r_df)
week9_df = week_reduce(week9r_df)
week10_df = week_reduce(week10r_df)
week11_df = week_reduce(week11r_df)
week12_df = week_reduce(week12r_df)
week13_df = week_reduce(week13r_df)
week14_df = week_reduce(week14r_df)
week15_df = week_reduce(week15r_df)
week16_df = week_reduce(week16r_df)
week17_df = week_reduce(week17r_df)
```

```python
# Merge the plays and games dataframes
full_play_df = pd.merge(plays_df, games_df, on = 'gameId', how = 'inner')
```

```python
# Add the defense column identifying the defensive team
full_play_df['def'] = np.where(full_play_df['possessionTeam'] != full_play_df['homeTeamAbbr'], 
                               full_play_df['homeTeamAbbr'], full_play_df['visitorTeamAbbr'])
```

```python
# Create a new column identifying offensive and defensive players
def weekly_conv(df):
    df['defPlayer'] = np.where((df['position'] == 'SS') | 
                                 (df['position'] == 'FS') |
                                 (df['position'] == 'MLB') | 
                                 (df['position'] == 'CB') |
                                 (df['position'] == 'LB') | 
                                 (df['position'] == 'OLB') |
                                 (df['position'] == 'ILB') | 
                                 (df['position'] == 'DL') |
                                 (df['position'] == 'DB') | 
                                 (df['position'] == 'NT') |
                                 (df['position'] == 'S') | 
                                 (df['position'] == 'DE') |
                                 (df['position'] == 'DT'),
                                 'D', 'O')
    
    return df.sort_values(by=['gameId', 'playId', 'frameId', 'team'])

w1_df = weekly_conv(week1_df)
w2_df = weekly_conv(week2_df)
w3_df = weekly_conv(week3_df)
w4_df = weekly_conv(week4_df)
w5_df = weekly_conv(week5_df)
w6_df = weekly_conv(week6_df)
w7_df = weekly_conv(week7_df)
w8_df = weekly_conv(week8_df)
w9_df = weekly_conv(week9_df)
w10_df = weekly_conv(week10_df)
w11_df = weekly_conv(week11_df)
w12_df = weekly_conv(week12_df)
w13_df = weekly_conv(week13_df)
w14_df = weekly_conv(week14_df)
w15_df = weekly_conv(week15_df)
w16_df = weekly_conv(week16_df)
w17_df = weekly_conv(week17_df)
```

```python
# Determine the defender closest to the football based on week, game, play and event
def get_closest_dplayer(wid, gid, pid, eid):
    
    # print(wid, gid, pid, eid)
    
    if wid == 1:
        w_df = w1_df
    elif wid == 2:
        w_df = w2_df
    elif wid == 3:
        w_df = w3_df
    elif wid == 4:
        w_df = w4_df
    elif wid == 5:
        w_df = w5_df
    elif wid == 6:
        w_df = w6_df
    elif wid == 7:
        w_df = w7_df
    elif wid == 8:
        w_df = w8_df
    elif wid == 9:
        w_df = w9_df
    elif wid == 10:
        w_df = w10_df
    elif wid == 11:
        w_df = w11_df
    elif wid == 12:
        w_df = w12_df
    elif wid == 13:
        w_df = w13_df
    elif wid == 14:
        w_df = w14_df
    elif wid == 15:
        w_df = w15_df
    elif wid == 16:
        w_df = w16_df
    elif wid == 17:
        w_df = w17_df
    
    # Test for Football entry
    err_test = w_df.loc[(w_df['displayName'] == 'Football') & 
                      (w_df['gameId'] == gid) &
                      (w_df['playId'] == pid) & 
                      (w_df['event'] == eid)]['x']
    
    # Return None of multiple copies exist
    if len(err_test) > 1:
        defender = "None"
        return defender
    
    # If Football is empty change event to pass_coutcome_touchdown
    if err_test.empty:
        eid = 'pass_outcome_touchdown'
    
    # Get ball position
    bpos_x = w_df.loc[(w_df['displayName'] == 'Football') & 
                      (w_df['gameId'] == gid) &
                      (w_df['playId'] == pid) & 
                      (w_df['event'] == eid)]['x']
    bpos_y = w_df.loc[(w_df['displayName'] == 'Football') & 
                      (w_df['gameId'] == gid) &
                      (w_df['playId'] == pid) &
                      (w_df['event'] == eid)]['y']
    
    # Create a frame df
    f_df = w_df.loc[(w_df['playId'] == pid) & (w_df['gameId'] == gid) & (w_df['event'] == eid)]
    
    # Initialize distance
    f_df['bd'] = 100.11
    
    # Calculate the distance
    for i in f_df.index:
        f_df.at[i, 'bd'] = math.sqrt(((bpos_x - f_df.at[i, 'x'])**2) + 
                                     ((bpos_y - f_df.at[i, 'y'])**2))
    
    d_df = f_df.loc[f_df['defPlayer'] == 'D']   
    d_df['rank'] = d_df['bd'].rank()
    if d_df.loc[d_df['rank'] == 1]['displayName'].empty:
        d_player = "None"
    else:
        d_player = d_df.loc[d_df['rank'] == 1]['displayName'].item()
    
    return d_player
```

```python
# Find the closest defender for each pass_outcome play
full_play_df['defender'] = "None"

for i in full_play_df.index:
    
    if full_play_df.at[i, 'passResult'] == 'C':
        full_play_df.at[i, 'defender'] = get_closest_dplayer(full_play_df.at[i, 'week'], 
                                                             full_play_df.at[i, 'gameId'],
                                                             full_play_df.at[i, 'playId'],
                                                             'pass_outcome_caught')
    elif full_play_df.at[i, 'passResult'] == 'I':
        full_play_df.at[i, 'defender'] = get_closest_dplayer(full_play_df.at[i, 'week'], 
                                                             full_play_df.at[i, 'gameId'],
                                                             full_play_df.at[i, 'playId'],
                                                             'pass_outcome_incomplete')
    elif full_play_df.at[i, 'passResult'] == 'IN':
        full_play_df.at[i, 'defender'] = get_closest_dplayer(full_play_df.at[i, 'week'], 
                                                             full_play_df.at[i, 'gameId'],
                                                             full_play_df.at[i, 'playId'],
                                                             'pass_outcome_interception')

full_play_df.head(20) 
```
