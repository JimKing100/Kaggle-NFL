```python
# Create a new column identifying offensive and defensive players
def weekly_conv(df):
    df['defPlayer'] = np.where((df['position'] == 'SS') | 
                                 (df['position'] == 'FS') |
                                 (df['position'] == 'MLB') | 
                                 (df['position'] == 'CB') |
                                 (df['position'] == 'LB') | 
                                 (df['position'] == 'OLB') |
                                 (df['position'] == 'ILB') | 
                                 (df['position'] == 'DL') |
                                 (df['position'] == 'DB') | 
                                 (df['position'] == 'NT') |
                                 (df['position'] == 'S') | 
                                 (df['position'] == 'DE') |
                                 (df['position'] == 'DT'),
                                 'D', 'O')
    
    return df.sort_values(by=['gameId', 'playId', 'frameId', 'team'])

w1_df = weekly_conv(week1_df)
w2_df = weekly_conv(week2_df)
w3_df = weekly_conv(week3_df)
w4_df = weekly_conv(week4_df)
w5_df = weekly_conv(week5_df)
w6_df = weekly_conv(week6_df)
w7_df = weekly_conv(week7_df)
w8_df = weekly_conv(week8_df)
w9_df = weekly_conv(week9_df)
w10_df = weekly_conv(week10_df)
w11_df = weekly_conv(week11_df)
w12_df = weekly_conv(week12_df)
w13_df = weekly_conv(week13_df)
w14_df = weekly_conv(week14_df)
w15_df = weekly_conv(week15_df)
w16_df = weekly_conv(week16_df)
w17_df = weekly_conv(week17_df)
```

```python
# Determine the closest defender to a receiver
def get_closest_dplayer(wid, gid, pid, eid, op, dp):
    
    # print(wid, gid, pid, eid)
    
    if wid == 1:
        w_df = w1_df
    elif wid == 2:
        w_df = w2_df
    elif wid == 3:
        w_df = w3_df
    elif wid == 4:
        w_df = w4_df
    elif wid == 5:
        w_df = w5_df
    elif wid == 6:
        w_df = w6_df
    elif wid == 7:
        w_df = w7_df
    elif wid == 8:
        w_df = w8_df
    elif wid == 9:
        w_df = w9_df
    elif wid == 10:
        w_df = w10_df
    elif wid == 11:
        w_df = w11_df
    elif wid == 12:
        w_df = w12_df
    elif wid == 13:
        w_df = w13_df
    elif wid == 14:
        w_df = w14_df
    elif wid == 15:
        w_df = w15_df
    elif wid == 16:
        w_df = w16_df
    elif wid == 17:
        w_df = w17_df
        
    # Test for multiple offensive player entries
    err_test_o = w_df.loc[(w_df['displayName'] == op) & 
                          (w_df['gameId'] == gid) &
                          (w_df['playId'] == pid) & 
                          (w_df['event'] == eid)]['x']
    
    # Test for multiple defensive player entries
    err_test_d = w_df.loc[(w_df['displayName'] == dp) & 
                          (w_df['gameId'] == gid) &
                          (w_df['playId'] == pid) & 
                          (w_df['event'] == eid)]['x']
    
    # Create a frame df
    f_df = w_df.loc[(w_df['playId'] == pid) & (w_df['gameId'] == gid) & (w_df['event'] == eid)]
    
    # Return None is non entries or multiple entries
    if (f_df.empty) | (err_test_o.count() > 1) | (err_test_d.count() > 1):
        return None, None
    
    # Determine offensive and defensive locations
    opos_x = f_df.loc[f_df['displayName'] == op]['x'].item()
    opos_y = f_df.loc[f_df['displayName'] == op]['y'].item()
    dpos_x = f_df.loc[f_df['displayName'] == dp]['x'].item()
    dpos_y = f_df.loc[f_df['displayName'] == dp]['y'].item()
    
    # Determine the time of the play
    p_time = f_df.loc[f_df['displayName'] == op]['time'].item()
    dt_string = p_time[11:23]
    dt_obj = datetime.datetime.strptime(dt_string, '%H:%M:%S.%f')
    
    # Calculate the distance between players
    dist = math.sqrt(((opos_x - dpos_x)**2) + 
                     ((opos_y - dpos_y)**2))
    
    return dist, dt_obj 

test, time = get_closest_dplayer(1, 2018090901, 1452, 'pass_forward', 'Julio Jones', 'Malcolm Jenkins')
print(test)
print(time)
```

```python
# For each play calculate the total distance and speed at start and end of pass (closeness)
for i in dist_df.index:
    wid = dist_df.at[i, 'weekId']
    gid = dist_df.at[i, 'gameId']
    pid = dist_df.at[i, 'playId']
    o_player = dist_df.at[i, 'oPlayer']
    d_player = dist_df.at[i, 'dPlayer']
    end_dist = dist_df.at[i, 'endDist']
    p_time = dist_df.at[i, 'endTime']
    dt_string = p_time[11:23]
    end_time = datetime.datetime.strptime(dt_string, '%H:%M:%S.%f')
    
    print(wid, gid, pid, o_player, d_player)
        
    start_dist,  start_time = get_closest_dplayer(wid, gid, pid, 'pass_forward', o_player, d_player)
    
    if (start_dist is not None):
        cov_dist = start_dist + end_dist
        speed = (start_dist - end_dist) / ((end_time - start_time).total_seconds())
        
        dist_df.at[i, 'startDist'] = start_dist
        dist_df.at[i, 'startTime'] = start_time
        dist_df.at[i, 'covDist'] = cov_dist
        dist_df.at[i, 'closeSpeed'] = speed
    else:
        dist_df.at[i, 'startDist'] = None
        dist_df.at[i, 'startTime'] = None
        dist_df.at[i, 'covDist'] = None
        dist_df.at[i, 'closeSpeed'] = None
```
