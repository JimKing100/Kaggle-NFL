```python
# Imports
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the data
plays_df = pd.read_csv('/Users/JKMacBook/Documents/Lambda/Kaggle/NFL/nfl-big-data-bowl-2021/full_play_2.csv')
dist_df = pd.read_csv('/Users/JKMacBook/Documents/Lambda/Kaggle/NFL/nfl-big-data-bowl-2021/distance_2.csv')
```

```python
# Clean up the columns
dist_df = dist_df.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1)
```

```python
# Calculate the starting distance, ending distance, coverage distance and closing speed
# for each targeted receiver/defender 
plays_df['route'] = None
plays_df['startDist'] = None
plays_df['endDist'] = None
plays_df['covDist'] = None
plays_df['closeSpeed'] = None

for i in plays_df.index:
    
    # print(i)
    
    route = dist_df.loc[(dist_df['dPlayer'] == plays_df.at[i, 'defender']) & 
                        (dist_df['gameId'] == plays_df.at[i, 'gameId']) &
                        (dist_df['playId'] == plays_df.at[i, 'playId'])]['route']
    
    if len(route) == 1:
    
        route = dist_df.loc[(dist_df['dPlayer'] == plays_df.at[i, 'defender']) & 
                            (dist_df['gameId'] == plays_df.at[i, 'gameId']) &
                            (dist_df['playId'] == plays_df.at[i, 'playId'])]['route'].item()
    
        startDist = dist_df.loc[(dist_df['dPlayer'] == plays_df.at[i, 'defender']) & 
                                (dist_df['gameId'] == plays_df.at[i, 'gameId']) &
                                (dist_df['playId'] == plays_df.at[i, 'playId'])]['startDist'].item()
    
        endDist = dist_df.loc[(dist_df['dPlayer'] == plays_df.at[i, 'defender']) & 
                              (dist_df['gameId'] == plays_df.at[i, 'gameId']) &
                              (dist_df['playId'] == plays_df.at[i, 'playId'])]['endDist'].item()
    
        covDist = dist_df.loc[(dist_df['dPlayer'] == plays_df.at[i, 'defender']) & 
                              (dist_df['gameId'] == plays_df.at[i, 'gameId']) &
                              (dist_df['playId'] == plays_df.at[i, 'playId'])]['covDist'].item()
    
        closeSpeed = dist_df.loc[(dist_df['dPlayer'] == plays_df.at[i, 'defender']) & 
                                 (dist_df['gameId'] == plays_df.at[i, 'gameId']) &
                                 (dist_df['playId'] == plays_df.at[i, 'playId'])]['closeSpeed'].item()
    
    
        plays_df.at[i, 'route'] = route
        plays_df.at[i, 'startDist'] = startDist
        plays_df.at[i, 'endDist'] = endDist
        plays_df.at[i, 'covDist'] = covDist
        plays_df.at[i, 'closeSpeed'] = closeSpeed
    
plays_df.head(20)
```

```python
plays_df.to_csv('/Users/JKMacBook/Documents/Lambda/Kaggle/NFL/nfl-big-data-bowl-2021/final/full_play_3.csv')
```
