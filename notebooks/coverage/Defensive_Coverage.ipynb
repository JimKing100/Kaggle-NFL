```python
# Create a new column identifying offensive and defensive players
def weekly_conv(df):
    df['defPlayer'] = np.where((df['position'] == 'SS') | 
                                 (df['position'] == 'FS') |
                                 (df['position'] == 'MLB') | 
                                 (df['position'] == 'CB') |
                                 (df['position'] == 'LB') | 
                                 (df['position'] == 'OLB') |
                                 (df['position'] == 'ILB') | 
                                 (df['position'] == 'DL') |
                                 (df['position'] == 'DB') | 
                                 (df['position'] == 'NT') |
                                 (df['position'] == 'S') | 
                                 (df['position'] == 'DE') |
                                 (df['position'] == 'DT'),
                                 'D', 'O')
    
    return df.sort_values(by=['gameId', 'playId', 'frameId', 'team'])

w1_df = weekly_conv(week1_df)
w2_df = weekly_conv(week2_df)
w3_df = weekly_conv(week3_df)
w4_df = weekly_conv(week4_df)
w5_df = weekly_conv(week5_df)
w6_df = weekly_conv(week6_df)
w7_df = weekly_conv(week7_df)
w8_df = weekly_conv(week8_df)
w9_df = weekly_conv(week9_df)
w10_df = weekly_conv(week10_df)
w11_df = weekly_conv(week11_df)
w12_df = weekly_conv(week12_df)
w13_df = weekly_conv(week13_df)
w14_df = weekly_conv(week14_df)
w15_df = weekly_conv(week15_df)
w16_df = weekly_conv(week16_df)
w17_df = weekly_conv(week17_df)
```

```python
# Determine the type of defensive coverage based on: 
#      1) The number of deep safeties
#      2) The CB position from the line of scrimmage (LOS)
#      3) The CB orientation relative to the receiver
def get_deep_count(wid, gid, pid, eid):
    
    # print(wid, gid, pid, eid)
    
    if wid == 1:
        w_df = w1_df
    elif wid == 2:
        w_df = w2_df
    elif wid == 3:
        w_df = w3_df
    elif wid == 4:
        w_df = w4_df
    elif wid == 5:
        w_df = w5_df
    elif wid == 6:
        w_df = w6_df
    elif wid == 7:
        w_df = w7_df
    elif wid == 8:
        w_df = w8_df
    elif wid == 9:
        w_df = w9_df
    elif wid == 10:
        w_df = w10_df
    elif wid == 11:
        w_df = w11_df
    elif wid == 12:
        w_df = w12_df
    elif wid == 13:
        w_df = w13_df
    elif wid == 14:
        w_df = w14_df
    elif wid == 15:
        w_df = w15_df
    elif wid == 16:
        w_df = w16_df
    elif wid == 17:
        w_df = w17_df
    
    m_or_z = 'None'
    scheme = 'None'
    
    # Test for Football entry
    err_test = w_df.loc[(w_df['displayName'] == 'Football') & 
                        (w_df['gameId'] == gid) &
                        (w_df['playId'] == pid) & 
                        (w_df['event'] == eid)]['x']
    
    # Return None if Football entry does not exist or multiple copies exist
    if ((len(err_test) > 1) | (err_test.empty)):
        return m_or_z, scheme
    
    # Get ball position
    bpos_x = w_df.loc[(w_df['displayName'] == 'Football') & 
                      (w_df['gameId'] == gid) &
                      (w_df['playId'] == pid) & 
                      (w_df['event'] == eid)]['x']
    
    # Create a frame df
    f_df = w_df.loc[(w_df['playId'] == pid) & (w_df['gameId'] == gid) & (w_df['event'] == eid)]
    
    # initialize the x position value and distance value
    f_df['bx'] = bpos_x
    f_df['bd'] = bpos_x
    
    # Calculate the x distance from the ball
    for i in f_df.index:
        f_df.at[i, 'bd'] = abs(f_df.at[i, 'x'] - bpos_x)
    
    # Count the number of defense players greater than 10 yards from LOS
    d_df = f_df.loc[f_df['defPlayer'] == 'D']
    deep_count = 0
    for i in d_df.index:
        if d_df.at[i, 'bd'] > 10:
            deep_count += 1
    
    # Count the CBs near the LOS (< 5 yards) and CBs facing the receiver 
    cb_df = f_df.loc[f_df['position'] == 'CB']
    cb_code = 0
    o_code = 0
    for i in cb_df.index:
        if cb_df.at[i, 'bd'] < 5:
            cb_code += 1
        if (((cb_df.at[i, 'o'] > 85) & (cb_df.at[i, 'o'] < 95)) |
            ((cb_df.at[i, 'o'] > 175) & (cb_df.at[i, 'o'] < 185))):
            o_code += 1
     
    if deep_count == 0:
        m_or_z = 'man'
        scheme = 'cover-0'
    elif deep_count == 1:
        if ((cb_code > 0) & (o_code > 0)):
            m_or_z = 'man'
            scheme = 'cover-1'
        else:
            m_or_z = 'zone'
            scheme = 'cover-3'
    elif deep_count == 2:
        if ((cb_code > 0) & (o_code > 0)):
            m_or_z = 'man'
            scheme = 'cover-2'
        else:
            m_or_z = 'zone'
            scheme = 'cover-4'
    
    return m_or_z, scheme
```

```python
full_play_df['manZone'] = "None"
full_play_df['scheme'] = 'None'

for i in full_play_df.index:
    
    full_play_df.at[i, 'manZone'], full_play_df.at[i, 'scheme'] = get_deep_count(full_play_df.at[i, 'week'],
                                                                                 full_play_df.at[i, 'gameId'],
                                                                                 full_play_df.at[i, 'playId'],
                                                                                 'ball_snap')

full_play_df.head(20) 
```
